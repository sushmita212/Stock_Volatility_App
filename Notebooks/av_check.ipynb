```python
import os
import requests
from dotenv import load_dotenv
from pathlib import Path
import pandas as pd
```

```python
env_path = Path("..") / "backend" / "app" / ".env"   # relative to Notebooks/
load_dotenv(dotenv_path=env_path)
```

```python
api_key = os.getenv("AV_API_KEY")
print(api_key)
```

```python
url = (
                "https://www.alphavantage.co/query?"
                "function=TIME_SERIES_DAILY&"
                f"symbol={'MSFT'}&"
                f"outputsize={'full'}&"
                "datatype=json&"
                f"apikey={api_key}"
            )
response = requests.get(url)
response_data = response.json()
```

```python
response_data.keys()
```

```python
response_data["Information"]
```

```python
pd.DataFrame(response_data['Time Series (Daily)']).T
```

```python
 # Read data into DataFrame (8.1.12 & 8.1.13)
stock_data = response_data['Time Series (Daily)']
df = pd.DataFrame.from_dict(stock_data, orient="index", dtype=float)

# Convert index to `DatetimeIndex` named "date" (8.1.14)
df.index = pd.to_datetime(df.index, format='%Y-%m-%d')
df.index.name = "date"

# Remove numbering from columns (8.1.15)
df.columns = [col.split('. ')[1] for col in df.columns]
```
