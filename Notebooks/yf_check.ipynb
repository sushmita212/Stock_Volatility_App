```python
import yfinance as yf
```

```python
def get_prices(symbol: str, period: str = "1y", interval: str = "1d"):
    df = yf.download(
        tickers=symbol,
        period=period,
        interval=interval,
        auto_adjust=True,
        progress=False,
        threads=False,
    )
    return df
```

```python
get_prices("AAPL")
```
